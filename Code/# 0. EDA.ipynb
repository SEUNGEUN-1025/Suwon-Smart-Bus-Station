```python
df_인구 = df_17.drop(['val'], axis=1)
df_인구['coordinates'] = df_인구['geometry'].apply(multipolygon_to_coordinates)
df_인구 = df_인구.merge(df_17.drop(['geometry'], axis=1), how='inner', on='gid')
df_인구 = df_인구.merge(df_18.drop(['geometry'], axis=1), how='inner', on='gid')
df_인구 = df_인구.merge(df_19.drop(['geometry'], axis=1), how='inner', on='gid')
df_인구 = df_인구.fillna(0)
df_인구['total_val'] = df_인구.iloc[:,3] + df_인구.iloc[:,4] + df_인구.iloc[:,5]
df_인구.columns = ['gid', 'geometry', 'coordinates', 'old_val', 'youth_val', 'young_val', 'total_val']

# 인구수 정규화
df_인구['old_normal'] = df_인구['old_val'] / np.max(df_인구['old_val'])
df_인구['youth_normal'] = df_인구['youth_val'] / np.max(df_인구['youth_val'])
df_인구['young_normal'] = df_인구['young_val'] / np.max(df_인구['young_val'])
df_인구['total_normal'] = df_인구['total_val'] / np.max(df_인구['total_val'])

# 100X100 grid기반 central point
df_인구['coord_cent'] = df_인구['geometry'].apply(lambda x : list(x[0].centroid.coords[0]))
df_인구['geo_cent'] = df_인구['coord_cent'].apply(lambda x : Point(x))

# ========= 인구 정보에 법정경계 코드 & 이름 부여 =========
# Step1. 중심점이 포함되는 법정구 / 법정동
# Step2. 100x100 면적이 포함되는 법정구 / 법정동

df_인구['EMD_CD'] = 0
df_인구['EMD_KOR_NM'] = 0
df_인구['SIG_CD'] = 0
df_인구['SIG_KOR_NM'] = 0

for i in range(len(df_인구)):
    for j in range(len(df_30)):
        # 법정동
        if df_30['geometry'][j].contains(df_인구['geo_cent'][i]):
            df_인구['EMD_CD'][i] = df_30['EMD_CD'][j]
            df_인구['EMD_KOR_NM'][i] = df_30['EMD_KOR_NM'][j]
        elif df_30['geometry'][j].intersects(df_인구['geometry'][i]):
            df_인구['EMD_CD'][i] = df_30['EMD_CD'][j]
            df_인구['EMD_KOR_NM'][i] = df_30['EMD_KOR_NM'][j]
        
df_인구 = df_인구[(df_인구['EMD_KOR_NM']!=0)] # 수원시 밖에 위치한 격자 -> 제거
df_인구 = df_인구.reset_index()

for i in range(len(df_인구)):
    for j in range(len(df_29)):
        # 법정구
        if df_29['geometry'][j].contains(df_인구['geo_cent'][i]):
            df_인구['SIG_CD'][i] = df_29['SIG_CD'][j]
            df_인구['SIG_KOR_NM'][i] = df_29['SIG_KOR_NM'][j]
        elif df_30['geometry'][j].intersects(df_인구['geometry'][i]):
            df_인구['SIG_CD'][i] = df_29['SIG_CD'][j]
            df_인구['SIG_KOR_NM'][i] = df_29['SIG_KOR_NM'][j]
```

```python
# 건물 코드와 종류에 대한 정보 가져오기
코드정의서 = pd.read_excel("input/33.코드정의서.xlsx")
building = 코드정의서[521:849].reset_index().drop('index', axis=1) # 건물용도코드만 분리
building.columns = ['BDTYP_CD','코드명']

# 건물 코드가 잘못 들어가 있는 경우 수정
df_27.loc[(df_27['BDTYP_CD']=='07199'),'BDTYP_CD']="06201"
df_27.loc[(df_27['BDTYP_CD']=='08203'),'BDTYP_CD']="08003"

# 건물 이름도 없고, 종류 또한 알 수 없어 제거
unknown = ["04505", "03024", "07108", "08800"]
df_27 = df_27[df_27['BDTYP_CD'].isin(unknown) == False].reset_index(drop=True)

# 건물 코드와 그에 따른 건물 종류(=코드명)부여
df_27['코드명']=0

for i in range(len(df_27)):
    for j in range(len(building)):
        if df_27['BDTYP_CD'][i]==building['BDTYP_CD'][j]:
            df_27['코드명'][i]=building['코드명'][j]

# 건물 종류별 데이터셋 생성

## 학교
school = df_27[df_27['코드명'].str.contains('학교')==True][['BDTYP_CD','BULD_NM','EMD_CD','코드명','geometry']].reset_index(drop=True)
## 병원
hospi = df_27[(df_27['코드명'].str.contains('병원')==True)|(df_27['코드명'].str.contains('한의원')==True)|(df_27['코드명'].str.contains('의원')==True)|(df_27['코드명'].str.contains('기타의료시설')==True)][['BDTYP_CD','BULD_NM','EMD_CD','코드명','geometry']].reset_index(drop=True)
## 주거용
house = df_27[(df_27['코드명'].str.contains('주택')==True)|(df_27['코드명'].str.contains('아파트')==True)|(df_27['코드명'].str.contains('고시원')==True)|(df_27['코드명'].str.contains('오피스텔')==True)][['BDTYP_CD','BULD_NM','EMD_CD','코드명','geometry']].reset_index(drop=True)
## 회사
company = df_27[(df_27['코드명'].str.contains('기타공공업무시설')==True)|(df_27['코드명'].str.contains('일반공공업무시설')==True)|(df_27['코드명'].str.contains('사무소')==True)|(df_27['코드명'].str.contains('기타일반업무시설')==True)|(df_27['코드명'].str.contains('기타사무소')==True)][['BDTYP_CD','BULD_NM','EMD_CD','코드명','geometry']].reset_index(drop=True)
## 공장
factory = df_27[(df_27['코드명'].str.contains('일반공장')==True)|(df_27['코드명'].str.contains('정비공장')==True)|(df_27['코드명'].str.contains('대형점')==True)][['BDTYP_CD','BULD_NM','EMD_CD','코드명','geometry']] .reset_index(drop=True)
## 상업 시설
store = df_27[(df_27['코드명'].str.contains('시장')==True)|(df_27['코드명'].str.contains('백화점')==True)|(df_27['코드명'].str.contains('대형판매점')==True)][['BDTYP_CD','BULD_NM','EMD_CD','코드명','geometry']].reset_index(drop=True)
## 관공서
public = df_27[(df_27['코드명'].str.contains('동사무소')==True)|(df_27['코드명'].str.contains('도서관')==True)|(df_27['코드명'].str.contains('우체국')==True)|(df_27['코드명'].str.contains('파출소')==True)|(df_27['코드명'].str.contains('경찰서')==True)|(df_27['코드명'].str.contains('소방서')==True)|(df_27['코드명'].str.contains('국가기관청사')==True)][['BDTYP_CD','BULD_NM','EMD_CD','코드명','geometry']].reset_index(drop=True)


# 각 건물의 중심점(=geo_cent)과 좌표(=coordinates)을 찾아주는 함수
def make_geocent(data):
    res = data['geometry'].apply(lambda x : list(x[0].centroid.coords[0]))
    res = res.apply(lambda x : Point(x))
    buffer = gpd.GeoSeries(res).buffer(0.003)
    coordinates = buffer.apply(polygon_to_coordinates)
    return [res, coordinates]

school['geo_cent'] = make_geocent(school)[0]
school['coordinates'] = school['geometry'].apply(multipolygon_to_coordinates)
hospi['geo_cent'] = make_geocent(hospi)[0]
hospi['coordinates'] = hospi['geometry'].apply(multipolygon_to_coordinates)
house['geo_cent'] = make_geocent(house)[0]
house['coordinates'] = house['geometry'].apply(multipolygon_to_coordinates)
company['geo_cent'] = make_geocent(company)[0]
company['coordinates'] = company['geometry'].apply(multipolygon_to_coordinates)
factory['geo_cent'] = make_geocent(factory)[0]
factory['coordinates'] = factory['geometry'].apply(multipolygon_to_coordinates)
store['geo_cent'] = make_geocent(store)[0]
store['coordinates'] = store['geometry'].apply(multipolygon_to_coordinates)
public['geo_cent'] = make_geocent(public)[0]
public['coordinates'] = public['geometry'].apply(multipolygon_to_coordinates)
```

```python
df_미세먼지 = pd.concat([df_미세먼지2019,df_미세먼지2020]).reset_index().drop(['index'],axis=1)
df_미세먼지 = df_미세먼지[df_미세먼지['측정일시각']!='2021.1.1 0:00']
df_미세먼지 = df_미세먼지.dropna(axis=0)
```

```python
df_미세먼지 = df_미세먼지.reset_index().drop(['index'],axis=1)
```

```python
df_미세먼지['측정일시각'] = pd.to_datetime(df_미세먼지['측정일시각']) 
df_미세먼지['year'] = df_미세먼지['측정일시각'].dt.year
df_미세먼지['month'] = df_미세먼지['측정일시각'].dt.month
```

```python
df_미세먼지년도별 = df_미세먼지.groupby(['측정소명','year']).mean().reset_index()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(font="Malgun Gothic", 
        rc={"axes.unicode_minus":False})

plt.rcParams.update({"figure.figsize" : (11, 6),
                     "axes.facecolor" : "white",
                     "axes.edgecolor":  "black"})

label = ['경수대로(동수원)', '고색동', '광교동', '신풍동', '영통동', '인계동', '천천동', '호매실동']

x = np.arange(len(label))

plt.bar(x-0.22,'미세먼지PM10농도값',data = df_미세먼지년도별[df_미세먼지년도별['year']==2019],color = "#8fba6c",edgecolor ="#8fba6c",width = 0.39)
plt.bar(x+0.22,'미세먼지PM10농도값',data = df_미세먼지년도별[df_미세먼지년도별['year']==2020],color = "#7092C1",edgecolor ="#7092C1",width = 0.39)
plt.xticks(x,label,fontsize=13)
plt.xlabel('',fontsize = 15)
plt.ylabel('',fontsize = 15)
plt.title('2019년 & 2020년 연평균 미세먼지 농도',fontsize = 20)
```

```python
cmap = sns.color_palette("husl", 8)
cmap.reverse()

plt.figure(figsize=(10,6))

sns.lineplot('일자','미세먼지PM10농도값',data = df_측정소월별, palette=cmap, hue = '측정소명',legend=None)
plt.title('2019년 & 2020년 월평균 미세먼지',fontsize=20)
#plt.legend(loc = 'upper right',fontsize=15)
plt.xticks(rotation=45,fontsize=13)
plt.xlabel('')
plt.ylabel('')
```
