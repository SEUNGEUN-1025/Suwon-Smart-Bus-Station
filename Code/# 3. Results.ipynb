# 1) 우선설치 스코어 생성
* 아래의 식으로 지수를 생성하여 상위 30개 정류소를 선정하였다

$우선설치 스코어= \sqrt{이용시간} *  {대기질지수 \over max(대기질지수)}$

```python
이용시간_pred=pd.read_csv('정류장별_이용시간예측.csv')
미세먼지_최종등급=pd.read_csv("미세먼지_최종등급.csv")
make_index=pd.merge(이용시간_pred, 미세먼지_최종등급, on='정류소ID', how='left')
make_index=make_index[['정류소ID','Label','최종등급']]
```

```python
make_index['scaled_ratio']=make_index['최종등급']/max(make_index['최종등급'])
make_index['final_index']=np.sqrt(make_index['Label'])*make_index['scaled_ratio']
df_01= pd.read_csv("input/1.수원시_버스정류장.csv")
df_01.rename(columns={'정류장ID':'정류소ID'}, inplace=True) 
df_final=make_index.sort_values(by="final_index", ascending=False).head(30)
df_final.rename(columns={'Label':'이용 시간 예측', '최종등급':'대기 질 예측','scaled_ratio':'대기 질 _scale'}, inplace=True)
df_final=pd.merge(df_final, df_01[['lon', 'lat','정류장명','정류소ID']],on='정류소ID',how='left')  
df_final.to_csv("df_final.csv", index=False)
```

# 2) 광고 Target층 도출
* 앞서 선정된 30개의 각 버스정류장의 반경 500m 내에 속하는 유동인구 수를 모두 더한 뒤, 버스정류장별로 가장 많은 성연령층을 광고 Target층으로 선정함

```python
df_final=pd.read_csv("df_final.csv")
df_15= pd.read_csv("input/15.수원시_성연령별_유동인구(2020).csv")
df_유동인구_성연령별=df_15.groupby(['lon', 'lat']).sum().reset_index().drop('STD_YM', axis=1)
```

```python
df_final['geo_cent'] = df_final.apply(lambda row : Point([row['lon'], row['lat']]), axis=1)
df_유동인구_성연령별['geo_cent'] = df_유동인구_성연령별.apply(lambda row : Point([row['lon'], row['lat']]), axis=1)

buffer_500= gpd.GeoDataFrame(df_final['geo_cent'], geometry='geo_cent')
buffer_500.crs = {'init':'epsg:4326'} # 좌표계가 지정되지 않아 알맞는 4326좌표계를 설정해줍니다.
buffer_500=buffer_500.to_crs({'init':'epsg:5179'})  # 4326좌표계는 미터단위 분석이 불가능하므로 500미터 버퍼 생성이 가능한 5179좌표계로 변환합니다.
buffer_500=buffer_500.buffer(500)
buffer_500=buffer_500.to_crs({'init':'epsg:4326'}) # 버퍼 후 4326좌표계로 되돌아옵니다.
```

```python
df_final['10대 남성']=0
df_final['20대 남성']=0
df_final['30대 남성']=0
df_final['40대 남성']=0
df_final['50대 남성']=0
df_final['60대 이상 남성']=0
df_final['10대 여성']=0
df_final['20대 여성']=0
df_final['30대 여성']=0
df_final['40대 여성']=0
df_final['50대 여성']=0
df_final['60대 이상 여성']=0
for i in range(len(df_final)):
    for j in range(len(df_유동인구_성연령별)):
        if buffer_500[i].contains(df_유동인구_성연령별.geo_cent[j]):
            df_final['10대 남성'][i]+=df_유동인구_성연령별['MAN_FLOW_POP_CNT_10G'][j]
            df_final['20대 남성'][i]+=df_유동인구_성연령별['MAN_FLOW_POP_CNT_20G'][j]
            df_final['30대 남성'][i]+=df_유동인구_성연령별['MAN_FLOW_POP_CNT_30G'][j]
            df_final['40대 남성'][i]+=df_유동인구_성연령별['MAN_FLOW_POP_CNT_40G'][j]
            df_final['50대 남성'][i]+=df_유동인구_성연령별['MAN_FLOW_POP_CNT_50G'][j]
            df_final['60대 이상 남성'][i]+=df_유동인구_성연령별['MAN_FLOW_POP_CNT_60GU'][j]
            df_final['10대 여성'][i]+=df_유동인구_성연령별['WMAN_FLOW_POP_CNT_10G'][j]
            df_final['20대 여성'][i]+=df_유동인구_성연령별['WMAN_FLOW_POP_CNT_20G'][j]
            df_final['30대 여성'][i]+=df_유동인구_성연령별['WMAN_FLOW_POP_CNT_30G'][j]
            df_final['40대 여성'][i]+=df_유동인구_성연령별['WMAN_FLOW_POP_CNT_40G'][j]
            df_final['50대 여성'][i]+=df_유동인구_성연령별['WMAN_FLOW_POP_CNT_50G'][j]
            df_final['60대 이상 여성'][i]+=df_유동인구_성연령별['WMAN_FLOW_POP_CNT_60GU'][j]
df_final['광고target']=str(0)
for i in range(len(df_final)):
    df_final['광고target'][i]=df_final.columns[9+np.argmax(df_final.iloc[i, 9:21])]
```

# 3) 최종 제출본 도출

```python
df_final=df_final[['정류소ID','정류장명','lon','lat','광고target']]
```
